```python
# Create a 20x21 grid: 20 columns of farmers + 1 column of librarians
rows = 20
cols = 21

# Initialize grid with farmer emoticons
grid = [['🧑‍🌾' for _ in range(cols)] for _ in range(rows)]

# Add a column of librarians at the end
for i in range(rows):
    grid[i][0] = '🧑‍🏫'  # librarian emoticon

# Display the grid with spacing
print('┌' + '─' * (cols * 3) + '┐')
for row in grid:
    # Add space between each emoticon
    print('│ ' + '  '.join(row) + ' │')
print('└' + '─' * (cols * 3) + '┘')

print(f"\nGrid size: {rows} x {cols}")
print(f"Farmers: {rows * (cols - 1)} | Librarians: {rows}")
```

```python
# Alternative: Display with HTML for larger emoticons in Jupyter
from IPython.display import HTML

# Create the same grid
rows = 20
cols = 21
grid = [['🧑‍🌾' for _ in range(cols)] for _ in range(rows)]

# Add librarians column
for i in range(rows):
    grid[i][0] = '🧑‍🏫'

# Create HTML with larger font size and circles
html = '<div style="font-size: 24px; line-height: 1.8; font-family: monospace;">'
for i, row in enumerate(grid):
    row_html = ''
    for j, cell in enumerate(row):
        # Add circle around the last 10 librarians (rows 10-19, column 0)
        if j == 0 and i >= 10:
            row_html += f'<span style="border: 2px solid red; border-radius: 50%; padding: 1px 1px; display: inline-block;">{cell}</span> '
        # Add circle around the last row of farmers (row 19, columns 1-20)
        elif i == 19 and j > 0:
            row_html += f'<span style="border: 2px solid red; border-radius: 50%; padding: 1px 1px; display: inline-block;">{cell}</span> '
        else:
            row_html += f'<span style="border: 0px; border-radius: 50%; padding: 2px 2px; display: inline-block;">{cell}</span> '
    html += row_html + '<br>'
html += '</div>'

HTML(html)
```
